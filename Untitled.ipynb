```python
def content_based_filtering(product_id):
    cat_subcat=pd.merge(category,subcategories,how='left',left_on='id',right_on='category_id')
    cat_subcat=cat_subcat[['category_id','name_x','id_y','name_y']]
    cat_subcat.columns=['category_id','category','subcategory_id','subcategory']
    specifications=pd.merge(specs[['id','name']],specsvalues[['spec_id','product_id']],left_on='id',right_on='spec_id')

    subcategory_id=addproduct[addproduct['id']==product_id]['subcategory_id'][0]
    items=addproduct[addproduct['subcategory_id']==subcategory_id]

    if not subcategory:
        category_id=addproduct[addproduct['id']==product_id]['category_id'][0]
        items=addproduct[addproduct['category_id']==category_id]
    product=pd.merge(items,cat_subcat[['category','subcategory_id','subcategory']],how='left',on='subcategory_id')
    product=product.drop(['image_1','image_2','image_3'],axis='columns')
    product=pd.merge(product,specifications,how='left',on='id')
    spec_list=product.groupby('id')['name_y'].apply(list)
    spec_str=[]
    for i in spec_list:
        spec_str.append('|'.join(map(str,i)))
    spec_str
    cleaned = items.set_index('id').specs.str.split('|', expand=True).stack()
    spec=pd.get_dummies(cleaned).groupby(level=0).sum()
    spec['product_id']=spec.index
    items=pd.merge(items,spec,left_on="id",right_on='product_id')
    
    return items.drop('product_id',axis='columns')['id']

content_based_filtering(1)
con.close()
```
